## Share Product Defaults

Create a `Product` class with class attributes `tax_rate` set to `0.08` and `currency` set to `"USD"`. Each product instance stores its own `name` and `price`.

The program creates `Product("Wireless Mouse", 29.99)` and `Product("Mechanical Keyboard", 89.99)`, then prints the shared class values through both instances and through the class itself.

```python
class Product:
    # TODO: set the shared tax rate to 0.08
    tax_rate = 0.08
    # TODO: set the shared currency to "USD"
    currency = "USD"

    def __init__(self, name, price):
        self.name = name
        self.price = price


mouse = Product("Wireless Mouse", 29.99)
keyboard = Product("Mechanical Keyboard", 89.99)

print(f"{mouse.name}: {mouse.tax_rate} {mouse.currency}")
print(f"{keyboard.name}: {keyboard.tax_rate} {keyboard.currency}")
print(f"Class defaults: {Product.tax_rate} {Product.currency}")
```

## Number Tickets with Class Counters

Issue support ticket labels using class attributes stored on `Ticket`: `prefix` is `"SUP"`, `next_number` starts at `1000`, and `issued_count` starts at `0`.

When a `Ticket` is created, store the current `Ticket.next_number` on the instance as `number`, then increase `Ticket.next_number` by `1` and `Ticket.issued_count` by `1`. The program creates tickets titled `"Cannot log in"`, `"Payment failed"`, and `"Export report"`.

```python
class Ticket:
    # TODO: set the shared ticket prefix to "SUP"
    prefix = "SUP"
    # TODO: start ticket numbering at 1000
    next_number = 1000
    # TODO: start the issued ticket count at 0
    issued_count = 0

    def __init__(self, title):
        self.title = title
        # TODO: assign this ticket's number from Ticket.next_number
        
        self.number = Ticket.next_number
        
        # TODO: advance Ticket.next_number and Ticket.issued_count
        Ticket.next_number += 1
        Ticket.issued_count += 1
        
    def label(self):
        return f"{Ticket.prefix}-{self.number}: {self.title}"


tickets = [
    Ticket("Cannot log in"),
    Ticket("Payment failed"),
    Ticket("Export report"),
]

for ticket in tickets:
    print(ticket.label())
print(f"issued: {Ticket.issued_count}")
print(f"next: {Ticket.next_number}")
```

## Observe a Shared Mutable Class Attribute

Solved
Demonstrate a shared mutable class attribute for a help desk log. Store one empty list on the `HelpDesk` class named `tickets`.

Each `HelpDesk` instance stores its own `name`. Its `add_ticket()` method appends the string `"<name>: <message>"` to the shared `tickets` list through `self.tickets`. The program creates `HelpDesk("North")` and `HelpDesk("South")`; North adds `"Printer jam"`, and South adds `"VPN down"`.

```python
class HelpDesk:
    # TODO: store one shared empty list on the class
    tickets = []

    def __init__(self, name):
        self.name = name

    def add_ticket(self, message):
        # TODO: append "<name>: <message>" to the shared list
        self.tickets.append(f"{self.name}: {message}")


north = HelpDesk("North")
south = HelpDesk("South")

north.add_ticket("Printer jam")
south.add_ticket("VPN down")

print(f"North view: {north.tickets}")
print(f"South view: {south.tickets}")
print(f"Same list: {north.tickets is south.tickets}")
```

## Trace Shadowing and Fallback

Trace instance-vs-class lookup for `Subscription.plan`. Start with the class attribute `plan` set to `"free"`, then create `Subscription("alice")` and `Subscription("bob")`.

After the first print, assign only `alice.plan` to `"pro"`. After printing the shadowed values and both instance dictionaries, change `Subscription.plan` to `"standard"`. After the next print, delete `alice.plan` so Alice falls back to the class attribute again.

```python
class Subscription:
    # TODO: set the shared starting plan to "free"
    plan = "free"

    def __init__(self, username):
        self.username = username


alice = Subscription("alice")
bob = Subscription("bob")

print(f"start: alice={alice.plan}, bob={bob.plan}, class={Subscription.plan}")

# TODO: give only alice an instance attribute plan set to "pro"
alice.plan = "pro"
print(f"after alice shadow: alice={alice.plan}, bob={bob.plan}, class={Subscription.plan}")
print(f"alice dict: {alice.__dict__}")
print(f"bob dict: {bob.__dict__}")

# TODO: change the class attribute plan to "standard"
Subscription.plan = "standard"
print(f"after class change: alice={alice.plan}, bob={bob.plan}, class={Subscription.plan}")

# TODO: delete alice's instance attribute named plan
del alice.plan
print(f"after delete: alice={alice.plan}, bob={bob.plan}, class={Subscription.plan}")
```

## Resolve Class Attributes Through the MRO

Use class attributes and the method resolution order. Set `BasePolicy.region` to `"global"`; set `JsonMixin.format` to `"json"` and `JsonMixin.retention_days` to `7`; set `RetentionMixin.retention_days` to `30`.

`AuditLog` inherits from `RetentionMixin`, `JsonMixin`, and `BasePolicy` in that order, has class attribute `level` set to `"INFO"`, and stores the instance attribute `event`. After creating `AuditLog("login")`, set only that instance's `level` to `"DEBUG"`, then set `AuditLog.region` to `"eu"` so it shadows the base class attribute.



```python
class BasePolicy:
    # TODO: set the base region to "global"
    region = "global"


class JsonMixin:
    # TODO: set the output format to "json"
    format = "json"
    # TODO: set this mixin's retention_days to 7
    retention_days = 7


class RetentionMixin:
    # TODO: set this mixin's retention_days to 30
    retention_days = 30


class AuditLog(RetentionMixin, JsonMixin, BasePolicy):
    # TODO: set the default log level to "INFO"
    level = "INFO"

    def __init__(self, event):
        self.event = event


log = AuditLog("login")
# TODO: shadow level on this instance with "DEBUG"
log.level = "DEBUG"
# TODO: define AuditLog.region as "eu"
AuditLog.region = "eu"

print("MRO:", " -> ".join(cls.__name__ for cls in AuditLog.__mro__[:4]))
print(f"retention_days: {log.retention_days}")
print(f"format: {log.format}")
print(f"log.level: {log.level}")
print(f"AuditLog.level: {AuditLog.level}")
print(f"region: {log.region}")
print(f"BasePolicy.region: {BasePolicy.region}")
print(f"log dict: {log.__dict__}")
```
