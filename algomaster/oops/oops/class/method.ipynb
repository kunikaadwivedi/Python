## Update a Wallet with Self

Track a wallet owned by "Mina". A Wallet starts with owner "Mina" and balance 20, receives 15, then spends 8.

The add and spend instance methods must update self.balance. The label instance method must read self.owner and self.balance and return a string in the form "Mina: $27".

```python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def add(self, amount):
        # TODO: increase this wallet's balance by amount.
        self.balance += amount

    def spend(self, amount):
        # TODO: decrease this wallet's balance by amount.
        self.balance -= amount

    def label(self):
        # TODO: return a label using this wallet's owner and balance.
        return f"{self.owner}: ${self.balance}"


wallet = Wallet("Mina", 20)
wallet.add(15)
wallet.spend(8)
print(wallet.label())
```

## Keep Cart State Separate

Keep separate shopping carts for "Ada" and "Lin". Ada's cart receives ("notebook", 4.50) and ("pen", 1.25); Lin's cart receives ("mug", 8.00).

The add_item method must append a (name, price) tuple to the receiving cart's self.items list. The total method must return the sum of prices in that same instance. The receipt_line method must return a string in the form "Ada: 2 items, total $5.75" using two digits after the decimal point.

```python
class ShoppingCart:
    def __init__(self, owner):
        self.owner = owner
        self.items = []

    def add_item(self, name, price):
        # TODO: store this item on this cart only.
        self.items.append((name, price))

    def total(self):
        # TODO: add the prices stored on this cart.
        return sum(item[1] for item in self.items)

    def receipt_line(self):
        # TODO: use this cart's owner, item count, and total.
        item_count = len(self.items)
        cart_total = self.total()
        return f"{self.owner}: {item_count} items, total ${cart_total:.2f}"


ada_cart = ShoppingCart("Ada")
lin_cart = ShoppingCart("Lin")

ada_cart.add_item("notebook", 4.50)
ada_cart.add_item("pen", 1.25)
lin_cart.add_item("mug", 8.00)

print(ada_cart.receipt_line())
print(lin_cart.receipt_line())
```

## Call One Method Three Ways

Represent a conference ticket for event 'PyCon' and seat "B12". The summary instance method must return exactly "PyCon seat B12" by reading self.event and self.seat.

Call the same method three ways already shown in the starter code: through the instance, through the class with the instance passed explicitly, and through a saved bound method.

```python
class Ticket:
    def __init__(self, event, seat):
        self.event = event
        self.seat = seat

    def summary(self):
        # TODO: use this ticket's event and seat.
        return f"{self.event} seat {self.seat}"


ticket = Ticket("PyCon", "B12")
bound_summary = ticket.summary

print(ticket.summary())
print(Ticket.summary(ticket))
print(bound_summary())
```
