## Read Class-Level Shipping Rates

Print the standard shipping label twice: once by calling a class method on `ShippingOption`, and once by calling the same class method through the `express` instance.

The class-level `standard_rate` is `4.99`. The `standard_shipping_label` class method should use `cls.standard_rate` and return a string formatted with exactly two decimal places.

```python
class ShippingOption:
    standard_rate = 4.99

    def __init__(self, name, rate):
        self.name = name
        self.rate = rate

    @classmethod
    def standard_shipping_label(cls):
        # TODO: Use cls.standard_rate to build the shipping label.
        return f"Standard shipping: ${cls.standard_rate:.2f}"


express = ShippingOption("Express", 12.50)

print(ShippingOption.standard_shipping_label())
print(express.standard_shipping_label())
```

## Build a Book from a Pipe-Separated Line

Create a `Book` instance from the exact string `"The Hobbit | J. R. R. Tolkien | 310"` using the `Book.from_line` class method.

The class method should split the line on the `"|"` character, strip whitespace from each part, convert the page count `"310"` to the integer `310`, and call `cls(title, author, pages)`.

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    @classmethod
    def from_line(cls, line):
        # TODO: Parse the line and return cls(title, author, pages).
        parts = line.split("|")
        title = parts[0].strip()
        author = parts[1].strip()
        pages = int(parts[2].strip())

        return cls(title, author, pages)


book = Book.from_line("The Hobbit | J. R. R. Tolkien | 310")

print(f"{book.title} by {book.author}")
print(f"{book.pages} pages")
```

## Manage Class-Specific Invoice Sequences

Create invoices with class-specific invoice number sequences. `Invoice.next_id` starts at `1000`, and `OnlineInvoice.next_id` starts at `5000`.

The `create` class method should build an invoice with `cls.next_id`, increment `cls.next_id` by `1`, and return the new invoice. The `reset_sequence` class method should set `cls.next_id` to the integer it receives. Use `cls` so `Invoice` and `OnlineInvoice` keep separate sequences.

```python
class Invoice:
    next_id = 1000

    def __init__(self, invoice_id, customer, total):
        self.invoice_id = invoice_id
        self.customer = customer
        self.total = total

    @classmethod
    def create(cls, customer, total):
        # TODO: Create an invoice using cls.next_id, then increment cls.next_id.
        invoice = cls(cls.next_id, customer, total)
        cls.next_id += 1
        return invoice

    @classmethod
    def reset_sequence(cls, next_id):
        # TODO: Set cls.next_id to next_id.
        cls.next_id = next_id
        


class OnlineInvoice(Invoice):
    next_id = 5000


invoices = [
    Invoice.create("Acme", 125.50),
    Invoice.create("Beta", 99.00),
    OnlineInvoice.create("WebCo", 49.95),
]

OnlineInvoice.reset_sequence(7000)
invoices.append(OnlineInvoice.create("ShopCo", 12.00))

for invoice in invoices:
    print(f"{type(invoice).__name__} #{invoice.invoice_id}: {invoice.customer} ${invoice.total:.2f}")

print(f"Invoice next: {Invoice.next_id}")
print(f"OnlineInvoice next: {OnlineInvoice.next_id}")
```
