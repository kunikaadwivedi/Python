## Write a Method That Uses self

Represent two menu items as `MenuItem` objects with instance attributes. The item named "Latte" costs 4.75, and the item named "Blueberry Muffin" costs 3.50.

Implement `display(self)` so it returns the instance's name and price formatted as `name - $price` with two decimal places, then call the method on both objects.

```python
class MenuItem:
    def display(self):
        # TODO: return this item's name and price using self.
        return f'{self.name} - ${self.price:.2f}'


latte = MenuItem()
latte.name = "Latte"
latte.price = 4.75

muffin = MenuItem()
muffin.name = "Blueberry Muffin"
muffin.price = 3.50

print(latte.display())
print(muffin.display())
```

## Keep Two Shopping Cart Instances Independent

Build two independent `ShoppingCart` objects for the store "Python Market". The class should have a class attribute `store_name` set to "Python Market" and an `add_item(self, item, price)` method that appends the item string to `self.items` and increases `self.total` by `price`.

Create `cart_a` for customer 101 with an empty item list and total 0.0, then add "mouse" for 29.99 and "cable" for 9.99. Create `cart_b` for customer 202 with an empty item list and total 0.0, then add "notebook" for 4.50.

```python
class ShoppingCart:
    # TODO: store the shared store name "Python Market" on the class.
    store_name = "Python Market"

    def add_item(self, item, price):
        # TODO: append item to self.items and add price to self.total.
        self.item = item
        self.items.append(item)
        self.price = price
        self.total += price


def cart_summary(cart):
    items_text = ", ".join(cart.items)
    return f"{cart.store_name} cart {cart.customer_id}: {items_text} = ${cart.total:.2f}"


cart_a = ShoppingCart()
# TODO: initialize cart_a for customer 101 with an empty item list and total 0.0.
cart_a.customer_id = 101
cart_a.items = []
cart_a.total = 0.0
# TODO: add "mouse" for 29.99 and "cable" for 9.99 to cart_a.
cart_a.add_item("mouse", 29.99)
cart_a.add_item("cable", 9.99)

cart_b = ShoppingCart()
# TODO: initialize cart_b for customer 202 with an empty item list and total 0.0.
cart_b.customer_id = 202
cart_b.items = []
cart_b.total = 0.0
# TODO: add "notebook" for 4.50 to cart_b.
cart_b.add_item("notebook", 4.50)

print(cart_summary(cart_a))
print(cart_summary(cart_b))
```

## Shadow a Class Attribute on One Instance

Create a `Customer` class with a class attribute `tier` equal to "standard" and a method `badge(self)` that returns the customer's name, tier, and points. Make one customer named "Alice" with 250 points and another named "Ben" with 480 points.

Set only Ben's instance attribute `tier` to "vip". Print both badges, then print the class-level tier, then print whether Alice's class is `Customer`, and finally print whether Alice and Ben are the same object.

```python
class Customer:
    # TODO: store the default tier "standard" on the class.
    tier = "standard"

    def badge(self):
        # TODO: return the customer's badge using self.name, self.tier, and self.points.
        return f'{self.name} [{self.tier}] - {self.points} pts'


alice = Customer()
# TODO: give Alice the name "Alice" and 250 points.
alice.name = "Alice"
alice.points = 250

ben = Customer()
# TODO: give Ben the name "Ben", 480 points, and the instance tier "vip".
ben.name = "Ben"
ben.points = 480
ben.tier = "vip"

print(alice.badge())
print(ben.badge())
print(Customer.tier)
print(alice.__class__ is Customer)
print(alice is ben)
```
