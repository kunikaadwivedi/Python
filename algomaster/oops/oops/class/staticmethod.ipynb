## Validate Product Prices

A product price is valid only when it is an int or float and is greater than or equal to 0. The class method should not read instance data or class data, so define the validator as a static method with one parameter named price.

The program checks the literal values 19.99, 0, -3, and "free" by calling Product.is_valid_price(...) and printing each result.

```python
class Product:
    # TODO: make this a static method.
    @staticmethod
    def is_valid_price(price):
        # TODO: return True only for int or float values greater than or equal to 0.
        
        return isinstance(price, (int, float)) and price >= 0


values = [19.99, 0, -3, "free"]
for value in values:
    print(Product.is_valid_price(value))
```
