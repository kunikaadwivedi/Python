## Store Product Details on One Instance

Create a `Product` instance for the product named "Wireless Mouse" priced at `29.99` with stock `50`.

Store the constructor values as instance attributes named `name`, `price`, and `stock` inside `__init__`, then read those attributes with dot syntax in the existing print calls.

```python
class Product:
    def __init__(self, name, price, stock):
        # TODO: store name, price, and stock as instance attributes.
        self.name = name
        self.price = price
        self.stock = stock


mouse = Product("Wireless Mouse", 29.99, 50)

print(f"Product: {mouse.name}")
print(f"Price: ${mouse.price:.2f}")
print(f"Stock: {mouse.stock}")
```

## Inspect Separate Instance Dictionaries

Create two `Product` objects with separate instance attribute dictionaries. The first product is named "Wireless Mouse" with price `29.99`; the second is named "Mechanical Keyboard" with price `89.99`.

Store `name` and `price` inside `__init__`. After construction, add `category = "peripherals"` and `tags = ["wireless", "ergonomic"]` only to the mouse instance, leaving the keyboard instance with only `name` and `price`.

```python
class Product:
    def __init__(self, name, price):
        # TODO: store name and price as instance attributes.
        self.name = name
        self.price = price


mouse = Product("Wireless Mouse", 29.99)
keyboard = Product("Mechanical Keyboard", 89.99) 

# TODO: add category and tags to mouse only.

mouse.category = 'peripherals'
mouse.tags = ['wireless', 'ergonomic']

print(mouse.__dict__)
print(keyboard.__dict__)
print(mouse.__dict__ is keyboard.__dict__)
```

## Mutate and Delete Account Attributes

Model bank account "A-100" with an opening balance of `120`. Store `account_id`, `balance`, and `status = "open"` as instance attributes when the account is constructed.

The `deposit` method increases `balance` by the given amount, the `withdraw` method decreases `balance` by the given amount, and the `close` method deletes the `status` instance attribute. The script deposits `35`, withdraws `20`, adds `note = "reviewed"` outside the class, and then closes the account.

```python
class BankAccount:
    def __init__(self, account_id, opening_balance):
        self.account_id = account_id
        # Using domain-friendly 'opening_balance' as the parameter, tracked as 'balance'
        self.balance = opening_balance
        self.status = "open"

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        # Decrease this account's balance
        self.balance -= amount

    def close(self):
        # Directly delete status since it is guaranteed to exist from __init__
        del self.status


account = BankAccount("A-100", 120)
account.deposit(35)
account.withdraw(20)
account.note = "reviewed"
account.close()

print(f"{getattr(account, 'account_id', '?')}: ${getattr(account, 'balance', 0)}")
print(f"Has status: {hasattr(account, 'status')}")
print(account.__dict__)
```

## Keep Mutable Instance Attributes Independent

Create two `Character` objects with independent `inventory` instance attributes. "Ada" starts with `['notebook']`, and "Ben" omits inventory so Ben receives a new empty list.

Store `name` as an instance attribute. Store a copy of any provided inventory list by calling `list(inventory)`, and create a new empty list when `inventory` is `None`. The `pick_up` method appends an item to that one character's inventory; the script adds "compass" to Ada and "lantern" to Ben.

```python
class Character:
    def __init__(self, name, inventory=None):
        # TODO: store name and an independent inventory list.
        self.name = name
        if inventory is None:
            self.inventory = []
        else:
            self.inventory = list(inventory)

    def pick_up(self, item):
        # TODO: append item to this character's inventory.
        self.inventory.append(item)


ada = Character("Ada", ["notebook"])
ben = Character("Ben")

ada.pick_up("compass")
ben.pick_up("lantern")

print(ada.__dict__)
print(ben.__dict__)
print(ada.inventory is ben.inventory)
```

## Restrict Instance Attributes with Slots

Create a slotted `Sensor` class whose instances allow exactly three instance attributes: `name`, `reading`, and `unit`. Store the sensor named "Boiler" with initial reading `98.6` and unit "C".

The `update` method changes only the `reading` instance attribute. After updating the reading to `101.2`, assigning `location = "Lab 4"` should be blocked because `location` is not listed in `__slots__`.



```python
class Sensor:
    # TODO: allow exactly name, reading, and unit as slotted instance attributes.
    __slots__ = ('name', 'reading', 'unit')

    def __init__(self, name, reading, unit):
        # TODO: store name, reading, and unit as instance attributes.
        self.name = name
        self.reading = reading
        self.unit = unit

    def update(self, new_reading):
        # TODO: update this sensor's reading.
        self.reading = new_reading


sensor = Sensor("Boiler", 98.6, "C")
sensor.update(101.2)

print(f"{getattr(sensor, 'name', '?')}: {getattr(sensor, 'reading', 0):.1f} {getattr(sensor, 'unit', '?')}") #getattr(object, attribute, default value)
print(f"Has __dict__: {hasattr(sensor, '__dict__')}") #The hasattr() function checks if an object has a specific attribute, returning True or False.

try:
    sensor.location = "Lab 4"
except AttributeError as error:
    print(f"Extra attribute blocked: {type(error).__name__}")
else:
    print("Extra attribute blocked: no")
```
