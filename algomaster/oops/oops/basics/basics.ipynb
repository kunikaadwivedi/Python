## Create a Product Class

Define a `Product` class that bundles three pieces of state: the string name `"Notebook"`, the float price `4.5`, and the integer stock count `12`.

The `__init__` method should store those values on `self.name`, `self.price`, and `self.stock`. The `label` method should return one formatted string using the stored attributes, with the price displayed to exactly 2 decimal places.

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price =price
        self.stock = stock
    
    def Label(self):
        return f'{self.name}: ${self.price:.2f} ({self.stock} in stock)'
   
product_a =Product("Notebook", 4.5, 12)
print(product_a.Label())
```

## Give Each Cart Its Own State

Build a `Cart` class where each instance stores its own customer name, item list, and running total. The cart for `"Alice"` receives `"Pen"` for `2.50` and `"Mug"` for `13.00`; the cart for `"Bob"` receives `"Tea"` for `7.25`.

The `summary` method should report the customer name, item count, and total formatted to 2 decimal places. The `item_names` method should return the cart's item names joined with `", "`.

```python
class Cart:
    def __init__(self, customer):
        self.customer = customer
        self.items = []
        self.total = 0.0

    def add_item(self, item, price):
        self.items.append(item)
        self.total += price

    def summary(self):
        return f'{self.customer}: {len(self.items)} items, total ${self.total:.2f}'

    def item_names(self):
        return ", ".join(self.items)


alice_cart = Cart("Alice")
alice_cart.add_item("Pen", 2.50)
alice_cart.add_item("Mug", 13.00)

bob_cart = Cart("Bob")
bob_cart.add_item("Tea", 7.25)

print(alice_cart.summary())
print(bob_cart.summary())
print(f"Alice items: {alice_cart.item_names()}")
print(f"Bob items: {bob_cart.item_names()}")
```

## Avoid Shared Mutable Class State

Create two independent `Playlist` objects named `"Road Trip"` and `"Focus"`. The `"Road Trip"` playlist gets the songs `"Go West"` and `"Open Road"`; the `"Focus"` playlist gets the song `"Deep Work"`.

Each playlist must create its own empty `self.songs` list inside `__init__`, so adding songs to one playlist does not change the other playlist. The `summary` method should use the playlist title and its own songs joined with `", "`.

```python
class Playlist:
    def __init__(self, title):
        self.title = title
        self.songs = []
        
    def add_song(self, song):
        self.songs.append(song)
        
    def summary(self):
        return f'{self.title}: {", ".join(self.songs)}'

road_trip = Playlist("Road Trip")
road_trip.add_song("Go West")
road_trip.add_song("Open Road")

focus = Playlist("Focus")
focus.add_song("Deep Work")

print(road_trip.summary())
print(focus.summary())
print(f"Different song lists: {road_trip.songs is not focus.songs}")
```

## Use Inheritance and Polymorphic Methods
Model notifications with one base class and two subclasses. `EmailNotification` should store the message plus the email address, and `SmsNotification` should store the message plus the phone number.

Create notification objects for `"Welcome Alice"` to `"alice@example.com"`, `"Code 4821"` to `"+15550123"`, and `"Receipt #1007"` to `"bob@example.com"`. The `send_all` function should call `send()` on each object in order, without checking the object's concrete class.

```python
from abc import ABC, abstractmethod
class Notification(ABC):
    def __init__(self, message):
        self.message = message
        
    @abstractmethod
    def send(self):
        pass
        
class EmailNotification(Notification):
    def __init__(self, message, email_address):
        super().__init__(message)
        self.email_address = email_address
        
    def send(self):
        return f"Email to {self.email_address}: {self.message}"
    
class SmsNotification(Notification):
    def __init__(self, message, phone_number):
        super().__init__(message)
        self.phone_number = phone_number
        
    def send(self):
        return f"SMS to {self.phone_number}: {self.message}"

def send_all(notifications):
    return [notification.send() for notification in notifications]

notifications = [
    EmailNotification("Welcome Alice", "alice@example.com"),
    SmsNotification("Code 4821", "+15550123"),
    EmailNotification("Receipt #1007", "bob@example.com"),
]

for line in send_all(notifications):
    print(line)
```

## Compose Objects for a Receipt

Represent a shopping receipt with three classes that work together. `Product` stores a name and price, `CartItem` stores a product and quantity, and `Cart` stores the customer name plus a list of cart items.

Create a cart for `"Nora"` containing 2 copies of `"Python 101"` priced at `29.95` and 3 copies of `"Code Mug"` priced at `12.50`. Each line item should compute its own total, and the cart should compute the receipt total from its line items.

```python
class Product:
    def __init__(self, name, price):
        # TODO: Store the product name and price.
        self.name = name 
        self.price = price


class CartItem:
    def __init__(self, product, quantity):
        # TODO: Store the product object and quantity.
        self.product = product
        self.quantity = quantity

    def line_total(self):
        # TODO: Return price times quantity for this item.
        return self.product.price * self.quantity

    def description(self):
        # TODO: Return one formatted receipt line for this item.
        return f'{self.product.name} x{self.quantity} = ${self.line_total():.2f}'


class Cart:
    def __init__(self, customer):
        # TODO: Store customer and create an empty item list.
        self.customer = customer
        self.items = []

    def add_product(self, product, quantity):
        # TODO: Add a CartItem for this product and quantity.
        self.items.append(CartItem(product, quantity))

    def total(self):
        # TODO: Return the sum of all item line totals.
        return sum(item.line_total() for item in self.items)

    def receipt(self):
        lines = [f"Receipt for {self.customer}"]
        lines.extend(item.description() for item in self.items)
        lines.append(f"Total: ${self.total():.2f}")
        return "\n".join(lines)


book = Product("Python 101", 29.95)
mug = Product("Code Mug", 12.50)
cart = Cart("Nora")
cart.add_product(book, 2)
cart.add_product(mug, 3)

print(cart.receipt())
```
