## Use Default Constructor Values

Build a SupportTicket constructor that always stores a title, a priority, and a resolved flag. The priority should default to the string "normal", and resolved should default to False when those arguments are not provided.

The program creates one ticket with only the title "Cannot log in" and another ticket with the title "Payment failed", priority "high", and resolved True.

```python
class SupportTicket:
    def __init__(self, title="", priority="normal", resolved=False):
        # TODO: make priority default to "normal" and resolved default to False.
        # TODO: store title, priority, and resolved on this instance.
        self.title = title
        self.priority = priority
        self.resolved = resolved


ticket_a = SupportTicket("Cannot log in")
ticket_b = SupportTicket("Payment failed", "high", True)

for ticket in (ticket_a, ticket_b):
    print(f"{ticket.title} | priority={ticket.priority} | resolved={ticket.resolved}")
```

## Validate Constructor Arguments

Create a BankAccount constructor that stores the owner and balance, but rejects any negative starting balance. If the balance is less than 0, __init__ must raise ValueError with the exact message "balance cannot be negative".

The program opens an account for "Maya" with 150.75 and then attempts to open an account for "Noah" with -20. The invalid account should be rejected by the constructor.

```python
class BankAccount:
    def __init__(self, owner, balance):
        # TODO: raise ValueError("balance cannot be negative") if balance is below 0.
        # TODO: store owner and balance on this instance.
        if balance < 0:
            raise ValueError("balance cannot be negative")
        self.owner = owner
        self.balance = balance


account = BankAccount("Maya", 150.75)
print(f"{account.owner}: ${account.balance:.2f}")

try:
    BankAccount("Noah", -20)
except ValueError as error:
    print(f"Rejected: {error}")
```

## Avoid Shared Mutable Defaults

Build a ShoppingCart constructor that stores an owner and an items list. If no items list is provided, create a new empty list for that specific cart; do not share one list between carts.

The program creates Alice's cart with ["notebook"] and Bob's cart with no starting items. It then adds "pen" to Alice's cart and "stapler" to Bob's cart, so each cart must keep only its own items.

```python
class ShoppingCart:
    def __init__(self, owner, items=None):
        # TODO: use None as the default value instead of a shared list.
        # TODO: store a new list for this instance, copying provided items.
        self.owner = owner
        if items is None:
            self.items = []
        else:
            self.items = list(items)

    def add_item(self, item):
        self.items.append(item)

alice_cart = ShoppingCart("Alice", ["notebook"])
bob_cart = ShoppingCart("Bob")

alice_cart.add_item("pen")
bob_cart.add_item("stapler")

print(f"{alice_cart.owner}: {', '.join(alice_cart.items)}")
print(f"{bob_cart.owner}: {', '.join(bob_cart.items)}")
print(f"Same list object: {alice_cart.items is bob_cart.items}")
```

## Assign Unique IDs During Initialization

Build a Subscription constructor that validates the plan before assigning an ID. Valid plans are exactly "Basic", "Pro", and "Team". An invalid plan must raise ValueError with the exact message "unknown plan: Enterprise" when the plan is "Enterprise".

For each valid instance, assign subscription_id from the class variable next_id, starting at 500, then increment next_id by 1. Store the customer, plan, monthly_price, and active flag on the instance; active should default to True.

```python
class Subscription:
    next_id = 500
    valid_plans = {"Basic", "Pro", "Team"}

    def __init__(self, customer, plan, monthly_price, active=True):
        # TODO: reject plans not in valid_plans before assigning an ID.
        # TODO: assign the current next_id, then increment next_id by 1.
        # TODO: store customer, plan, monthly_price, and active on this instance.
        if plan not in Subscription.valid_plans:
            raise ValueError(f"unknown plan: {plan}")
        
        self.subscription_id = Subscription.next_id
        Subscription.next_id += 1
        
        self.customer = customer
        self.plan = plan
        self.monthly_price = monthly_price
        self.active = active


subscriptions = [
    Subscription("Alice", "Basic", 9.99),
    Subscription("Ben", "Pro", 19.99),
    Subscription("Cara", "Team", 49.50, False),
]

for subscription in subscriptions:
    print(
        f"#{subscription.subscription_id} {subscription.customer} "
        f"{subscription.plan} ${subscription.monthly_price:.2f} "
        f"active={subscription.active}"
    )

try:
    Subscription("Dana", "Enterprise", 99.00)
except ValueError as error:
    print(f"Rejected: {error}")

print(f"Next id: {Subscription.next_id}")
```
